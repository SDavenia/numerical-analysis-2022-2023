```python
import scipy
from scipy.special import binom

def bernstein(x, i, n):
    return binom(n-1, i)*x**i*(1-x)**(n-i-1)
```

```python
x = linspace(0,1,256)
```

```python
n = 5

for i in range(n):
    plot(x, bernstein(x,i,n))
```

```python
V = array([bernstein(x, i, n) for i in range(n)]).T
V.shape
X = linspace(0,1,n)
```

```python
def my_fun(x):
    return sin(2*pi*x)


plot(x, my_fun(x))
plot(x, V.dot(my_fun(X)))
```

```python
def bernstein_approx(x, u, n):
    V = array([bernstein(x, i, n) for i in range(n)]).T
    X = linspace(0,1,n)
    return V.dot(u(X))
```

```python
def abs_value(x):
    return abs(x-.5)
plot(x, abs_value(x))
_ = [plot(x, bernstein_approx(x, abs_value, i)) for i in range(10, 200,10)]
```
